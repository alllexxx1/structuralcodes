```python
import structuralcodes as sc
import matplotlib.pyplot as plt
import numpy as np
```

```python
el = sc.core.constitutive_laws.Elastic(210000)
strain = np.linspace(0,0.05,200)
stress = np.zeros_like(strain)
for i in range(len(strain)):
    stress[i] = el.getStress(strain[i])

plt.plot(strain,stress)
```

```python
el = sc.core.constitutive_laws.ElasticPlastic(210000,450,eps_su = 0.04)
strain = np.linspace(0,0.05,200)
stress = np.zeros_like(strain)
for i in range(len(strain)):
    stress[i] = el.getStress(strain[i])

plt.plot(strain,stress)
```

```python
el = sc.core.constitutive_laws.ParabolaRectangle(fc=-35)
strain = np.linspace(0,-0.004,200)
stress = np.zeros_like(strain)
for i in range(len(strain)):
    stress[i] = el.getStress(strain[i])

plt.plot(strain,stress)
```
